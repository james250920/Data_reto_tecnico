## Fase 2 — Limpieza y Transformación

A continuación se aplican las transformaciones necesarias para obtener un dataset limpio y consistente. Cada paso está justificado.

```python
datase = datase[datase['Billing Amount'] >= 0].copy()
datase.reset_index(drop=True, inplace=True)
datase['Billing Amount'].describe()
```

> **Nota:** Los valores bajos en 'Billing Amount' pueden corresponder a pagos de medicamentos u otros servicios menores, por lo que no se eliminan automáticamente del dataset. Esta decisión se toma para evitar perder información potencialmente relevante para el análisis.

**Justificación:** Se eliminan registros con 'Billing Amount' negativo porque no tienen sentido en el contexto hospitalario. Se resetea el índice para evitar inconsistencias.

```python
import re
def limpiar_prefijo(nombre):
    if isinstance(nombre, str):
        return re.sub(r'^(mr\.?|mrs\.?|ms\.?|miss|dr\.?|prof\.?|sr\.?|sra\.?|mx\.?|mtro\.?|lic\.?|ing\.?|dra\.?|fr\.?|rev\.?|capt\.?|col\.?|maj\.?|lt\.?|sgt\.?|cpt\.?|cmdr\.?|adm\.?|ofc\.?|pres\.?|gov\.?|sen\.?|rep\.?|amb\.?|treas\.?|sec\.?|pvt\.?|hon\.?|judge|arq\.?|qfb\.?|mc\.?|phd\.?|ph\.d\.?|m\.d\.?|d\.d\.s\.?|dvm\.?|esq\.?|srta\.?|srto\.?|licda\.?|licdo\.?|licenciado|licenciada|doctor|doctora|ingeniero|ingeniera|arquitecto|arquitecta|profesor|profesora|fray|hermano|hermana|padre|monja|reverendo|reverenda|capitán|coronel|mayor|teniente|sargento|comandante|almirante|oficial|presidente|gobernador|senador|representante|embajador|tesorero|secretario|privado|honorable|juez)\s+', '', nombre, flags=re.IGNORECASE).strip()
    return nombre
datase.loc[:, 'Name'] = datase['Name'].apply(limpiar_prefijo)
datase[['Name']]
```

**Justificación:** Se limpian los prefijos de títulos en los nombres para estandarizar la columna 'Name'.

```python
datase.info()
```

```python
import re

def limpiar_and(nombre):
    if isinstance(nombre, str):
        return re.sub(r'^(And\s+)|(\s+And)$', '', nombre, flags=re.IGNORECASE).strip()
    return nombre
datase['Hospital'] = datase['Hospital'].apply(limpiar_and)

datase['Hospital'].unique()[:10]  
```

**Justificación:** Se limpian inconsistencias en los nombres de hospitales para facilitar análisis posteriores.

```python

def limpiar_prefijo(nombre):
    import re
    if isinstance(nombre, str):
        return re.sub(r'^(mr|mrs|ms|dr|miss|sr|sra|srta)\.?\s+', '', nombre, flags=re.IGNORECASE).strip()
    return nombre

datase['Name'] = datase['Name'].apply(limpiar_prefijo)

datase['Name'].head(10) 
```
